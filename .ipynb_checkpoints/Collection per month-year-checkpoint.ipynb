```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("grossing2017-2011.csv", parse_dates=['estreia'])
```

```python
renda = dataset[['renda','estreia']]#renda por data de estreia
```

```python
renda = renda.set_index('estreia')
renda = renda.groupby(lambda x: (x.year, x.month)).sum()#cotablizacao por mes do ano
```

```python
renda.reset_index(inplace=True)
```

```python
renda.rename({'index': 'date'}, axis=1, inplace=True)
```

```python
renda.plot(x= 'date', y='renda')
```
