```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("dataset_full.csv", index_col=False)
```

```python
dataset.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
dataset
```

```python
dataset['distribuidora'] = dataset['distribuidora'].str.lower()
```

```python
dataset
```

```python
from sklearn import preprocessing
```

```python
# criando uma instancia do label encoder
le = preprocessing.LabelEncoder()
```

```python
distLE = preprocessing.LabelEncoder()
```

```python
generoLE = le.fit_transform(dataset['genero'])
subgeneroLE = le.fit_transform(dataset['subgenero'])
```

```python
distribuidora = distLE.fit_transform(dataset['distribuidora'])
```

```python
dataset['genero'] = generoLE
dataset['subgenero'] = subgeneroLE
```

```python
dataset['distribuidora'] = distribuidora
```

```python
distLE.classes_
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
```

```python
lreg = LinearRegression()
```

```python
labels = dataset['renda']
train1 = dataset.drop(['renda', 'estreia', 'título_original', 'título_brasileiro', 'distribuidora'], axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size=0.20, random_state=34)
```

```python
lreg.fit(x_train, y_train)
```

```python
lreg.score(x_test, y_test)
```
