```python
dataset = pd.read_csv("grossing2017-2011.csv")
```

```python
dataset = dataset.drop_duplicates(subset='título_brasileiro', keep="first")
```

```python
dataset.count()
```

## Pytrends

```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360) #CONECTANDO COM O GOOGLE
```

```python
def sixMonthsEarlier(releaseDay):
    data = datetime.date.fromisoformat(releaseDay)
    sixMonthsAgo = datetime.timedelta(days=180)
    return str(data - sixMonthsAgo)

def getTimeFrame(i):
    to_tf = dataset.loc[i,'estreia']
    from_tf =  sixMonthsEarlier(to_tf)
    tf = str(from_tf +' '+ to_tf)
    tf = str(from_tf + " " + to_tf)
    return tf
```

```python
lista_teste = list(dataset['título_brasileiro'])
```

### tratando termos de pesquisa

```python
for i in range(len(lista_teste)):
    lista_teste[i] = lista_teste[i].replace(" -", ":")
```

```python
for i in range(len(lista_teste)):
    lista_teste[i] = lista_teste[i].replace("!", "")
```

```python
for i in range(len(lista_teste)):
    lista_teste[i] = lista_teste[i].replace("?", "")
```

```python
lista_teste
```

#### abaixo, filmes que a API não conseguiu pesquisar, por causa da grafia provavelmente

```python
filmes = []
for i in range(len(lista_teste)):
    try:
        lista_temp = [lista_teste[i]]
        pytrends.build_payload(lista_temp, timeframe = getTimeFrame(i), geo='BR')
        interest = pytrends.interest_over_time()
        filmes.append(interest.reset_index().drop(['date', 'isPartial'], axis=1).transpose())
    except:
        print(lista_teste[i])
        continue
```

```python
df = pd.concat(filmes)
df.reset_index(inplace=True)
df.rename({'index':'título_brasileiro'}, axis=1, inplace=True)
```

```python
df
```

```python
df_filmes = pd.merge(dataset, df, on='título_brasileiro')
```

```python
df_filmes.to_csv('filmes_v1.csv', index=False)
```

### Unindo Genero ao dataset

```python
dataset_filmes = pd.read_csv("filmes_v1.csv")
```

```python
filmes_genero = pd.read_csv("filmes_genero.csv")
```

```python
dataset_full = pd.merge(filmes_genero, dataset_filmes, on='título_original')
```

```python
dataset_full.drop("id", axis=1, inplace=True)
```

```python
dataset_full.to_csv('dataset_full.csv')
```

```python
dataset_full["0-10"] = 0
dataset_full["11-20"] = 0
dataset_full["21-30"] = 0
dataset_full["31-40"] = 0
dataset_full["41-50"] = 0
dataset_full["51-60"] = 0
dataset_full["61-70"] = 0
dataset_full["71-80"] = 0
dataset_full["81-90"] = 0
dataset_full["91-100"] = 0
```

```python
dataset_full
```

```python
# catalogar dias numa casa de 10
for i in range(len(dataset_full)):
    ranges = dataset_full.loc[i, '0':'180']
    dataset_full.loc[i, '0-10'] = sum(ranges.between(0,10))
    dataset_full.loc[i, '11-20'] = sum(ranges.between(11,20))
    dataset_full.loc[i, '21-30'] = sum(ranges.between(21,30))
    dataset_full.loc[i, '31-40'] = sum(ranges.between(31,40))
    dataset_full.loc[i, '41-50'] = sum(ranges.between(41,50))
    dataset_full.loc[i, '51-60'] = sum(ranges.between(51,60))
    dataset_full.loc[i, '61-70'] = sum(ranges.between(61,70))
    dataset_full.loc[i, '71-80'] = sum(ranges.between(71,80))
    dataset_full.loc[i, '81-90'] = sum(ranges.between(81,90))
    dataset_full.loc[i, '91-100'] = sum(ranges.between(91,100))
```
