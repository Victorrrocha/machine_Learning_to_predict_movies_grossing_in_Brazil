## Pytrends

```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360) #CONECTANDO COM O GOOGLE
```

```python
movies = dataset.iloc[:,1]
movies.head()
```

```python
def sixMonthsEarlier(releaseDay):
    data = datetime.date.fromisoformat(releaseDay)
    sixMonthsAgo = datetime.timedelta(days=180)
    return str(data - sixMonthsAgo)

def getTimeFrame(i):
    to_tf = dataset.loc[i,'estreia']
    from_tf =  sixMonthsEarlier(to_tf)
    tf = str(from_tf +' '+ to_tf)
    tf = str(from_tf + " " + to_tf)
    return tf
```

```python
lista_teste = ['Velozes e furiosos 8','Liga da Justiça','A Bela e a Fera','Meu malvado favorito 3','Mulher-Maravilha']
```

```python
for i in range(5):
    lista_temp = [lista_teste[i]]
    print(lista_temp)
```

```python
filmes = []
for i in range(5):
    lista_temp = [lista_teste[i]]
    pytrends.build_payload(lista_temp, timeframe = getTimeFrame(i), geo='BR')
    interest = pytrends.interest_over_time()
    filmes.append(interest.reset_index().drop(['date', 'isPartial'], axis=1).transpose())
```

```python
df = pd.concat(filmes)
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.rename({'index':'título_brasileiro'}, axis=1, inplace=True)
```

```python
df.head()
```

```python
df_filmes = pd.merge(dataset, df, on='título_brasileiro')
```

```python
df_filmes.to_csv('filmes_v1.csv', index=False)
```
